```python
import pandas as pd
from sqlalchemy import create_engine

def create_headers_and_types(columns):
    headers = [column for column, _ in columns]
    types = {column: data_type.__name__ for column, data_type in columns}
    return headers, types

def write_dataframe_to_sql_table(df, name, schema, engine):
    try:
        # Write DataFrame to SQL table
        df.to_sql(name=name, schema=schema, con=engine, if_exists="replace", index=False)
        print(f"DataFrame successfully written to SQL table '{name}' in schema '{schema}'.")
    except Exception as e:
        print(f"Error writing DataFrame to SQL table '{name}' in schema '{schema}': {e}")

# Define your SQL Server connection details
server = "CLOUD-39YWHX3"
database = "psp_cr"
driver = "ODBC Driver 17 for SQL Server"
connection_string = f"mssql://@{server}/{database}?driver={driver}&trusted_connection=yes"

# Create SQLAlchemy engine
engine = create_engine(connection_string)

mapping_file_path = "C:\\Users\\Jiri.Pillar\\Developer\\Python\\poslanecka_snemovna_cr\\headers\\mapping.csv"

# Load data mapping
mapping_df = pd.read_csv(mapping_file_path, header=0)

for index, row in mapping_df.iterrows():
    schema = row['schema']
    table = row['table']
    file_name = row['file_name']

    file_path = f"input\\{file_name}"
    
    # Load mapping for headers and types
    headers, types = create_headers_and_types(tables[table])

    # Load data with specified headers and types
    df = pd.read_csv(
        file_path,
        encoding="windows-1250",
        sep="|",
        header=None,
        names=headers,
        index_col=False,
        low_memory=False,
    )
    
    # Fill NaN values with 0
    df.fillna(0, inplace=True)
    
    # Convert data types
    for column, data_type in types.items():
        df[column] = df[column].astype(data_type)
    
    # Write DataFrame to SQL table
    write_dataframe_to_sql_table(df, table, schema, engine)
```
